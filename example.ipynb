```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import h5py
from tqdm.notebook import tqdm
import multiprocessing as mp
```

```python
root = '/home/as1530/'

step = 20

def load():
    with h5py.File(f'{root}flds.tot.{step:05d}', 'r') as f:
        data = {}
        for d in ['bx', 'bz', 'dens1', 'dens2', 'dens3', 'dens7', 'dens8', 'dens9', 'xx', 'zz']:
            data[d] = f[d][:]
        return data

data = load()
```

```python
def plot(f, y0):
    take_slice = (slice(None), y0, slice(None))
    density = np.sum([f[f'dens{i}'][take_slice] for i in [1, 2, 3, 7, 8, 9]], 0)
    bx = f['bx'][take_slice]
    bz = f['bz'][take_slice]
    x = f['xx'][take_slice][0]
    z = f['zz'][take_slice][:, 0]
    xmin, xmax, zmin, zmax = x.min(), x.max(), z.min(), z.max()
    plt.figure(figsize=(10, 4), dpi=100, facecolor="w")
    plt.imshow(density,
        cmap='turbo', norm=mpl.colors.Normalize(0, 20),
        origin='lower', extent=(xmin, xmax, zmin, zmax))
    plt.streamplot(x, z, bx, bz,
        density=2, color='w',
        linewidth=0.1, arrowsize=1, arrowstyle='->')
    plt.xlabel('x [cells]')
    plt.ylabel('z [cells]')
    plt.xlim(xmin, xmax)
    plt.ylim(zmin, zmax)
    plt.tight_layout()

# plot(data, 10)

def plotAndSave(i):
    plot(data, i)
    plt.savefig(f'pics/fig{i:05d}.png')
    plt.close()

pool = mp.Pool(mp.cpu_count())
pool.map(plotAndSave, range(500))

# for i, y in enumerate(tqdm(range(0, 500, 5))):
#     plot(data, y)
#     plt.savefig(f'pics/fig{i:05d}.png')
#     plt.close()

# ffmpeg -r 10 -i folder/fig%d.png -c:v libx264 -pix_fmt yuv420p folder.mp4
```
